```python
from sklearn.model_selection import train_test_split
```

```python
hr.shape
```

```python
hr.shape[0]*.7
```

```python
np.random.seed(100)
train_rows = np.random.randint(1,hr.shape[0], int(hr.shape[0]*.7))
train_rows
```

```python
# MANUAL CODE WHICH IS NOT NEEDED
#splitting it into train
train = hr.iloc[train_rows]
print(train.shape)
train.head()
```

##  First Step: Get dummies

```python
hr_dummies = pd.get_dummies(hr)
hr_dummies
# dummies will create separate columns for each value of the categories we have
'''for example in the iris dataset we had 3 values inspecies column namely setosa, virgininca and versicolor
    so, when we get dummies for iris we get species.setosa, species.virginica, species.versicolor, 
    also the no of rows will be same and the flag for the respective category will be 1, and 0 for rest'''
```

##  Second Step: Split data into test and train

```python
# One line method to split the data using sklearn
train, test = train_test_split(hr_dummies,
                               test_size = 0.3,
                               random_state=100)
print(train.shape)
print(test.shape)
```

##  Segregate Data

```python
# Segregating input and output
train_y = train['Attrition']  # target column, i.e output or labels
train_x = train.drop('Attrition', axis=1)# every column except target column, i.e input columns or features

test_y = test['Attrition']   # target column, i.e output or labels 
test_x = test.drop('Attrition', axis=1) # every column except target column, i.e input columns or features

train_x.shape
```

```python
print(train_y.shape)
print(test_y.shape)
print(train_x.shape)
print(test_x.shape)
```

## Building Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Creating a model
model = DecisionTreeClassifier()
model.fit(train_x, train_y)
```

```python
# predicting on test data
test_pred = model.predict(test_x)
print(len(test_pred))
test_pred
```

```python
df_pred = pd.DataFrame({'actual': test_y,
                         'predicted': test_pred})
df_pred.head()
```

```python
df_pred['Status'] = df_pred['actual'] == df_pred['predicted']
df_pred.head()
```

```python
df_pred['Status'].value_counts() / df_pred.shape[0] * 100
```

##  Machine Learning Process
- Clean/Prepare your data (EDA Process)
    - Missing Value Treatment
    - Outlier removal
    - Standardization
    - Dummy variable Conversion
- Split your data in training and testing
- Separate input & output columns (train_x, train_y, test_x, test_y)
- Build model using train data
- predict output values for test data
- Evalueate your model
    - Accuracy ,sensitivity, specificity
- Fine tune your model(s) for better performance
    - Hyper parameter tuning
    - cross validation
