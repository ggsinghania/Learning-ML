## suppose user a has watched some 3 movies 
## like wise many users will 5 have watched some other movies 
## we would take the similarity of the neighbors 
### and the compare the movies watched with that of the neighbors having the highest cosine similarity 
## say , a user matches high value of cosine similarity with that of the c

## then we can recommend the movies with sorting in descending order 


```python
##we recommend the movies to the user 100
## we would go till the neighbors 7
## procedure to fill the nans
## row wise mean is 0 not column wise 
##so we would add the ratings n then we would subtract it from the mean 
## 
```

```python

ui_matrix = user_itme_matrix.apply(lambda v: v - v.mean(), axis=1).fillna(0)

ui_matrix.head()
```

```python
sim_matrix = pd.DataFrame(cosine_similarity(ui_matrix),
                          index = ui_matrix.index,
                          columns= ui_matrix.index)
```

```python
neighs_user = sim_matrix['User-100'].drop('User-100').sort_values(ascending = False).head(7).index
```

```python
def ubcf(userid):
    preds = pd.DataFrame(columns=['mid','predicted_rating'])
    neighs_user = sim_matrix[userid].drop(userid).sort_values(ascending = False).head(7).index
    for mid, rating in user_itme_matrix.loc[userid].items():
        if pd.isnull(rating):
            
            mid_ratings = user_itme_matrix.loc[neighs_user, mid]
            avg_mid_rating = mid_ratings.mean()
            preds = preds.append({'mid': mid,
                                  'predicted_rating': avg_mid_rating},
                                ignore_index = True)
            #print(mid, mid_ratings.values, avg_mid_rating)


    preds['movieId'] = preds['mid'].str.replace('Movie-',
                                                '').astype(int)

    preds = pd.merge(preds, movies, left_on = 'movieId', right_on='movieId')

    preds.sort_values('predicted_rating',
                      ascending = False).head(3)['title']



    movies_recom = preds.sort_values('predicted_rating', ascending = False).head(3)
    return movies_recom.tolist()
```
